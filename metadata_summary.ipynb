```python
counter = 0
data = pd.read_csv(os.path.join(isic_dataset_dir, 'truth.csv'))

age = data["age_approx"]
sex = data["sex"]
anatomy = data["anatom_site_general_challenge"]
benign_malignant = data["benign_malignant"]

age = px.histogram(data, x="age_approx")
age.update_layout(
    xaxis_title="",
    yaxis_title="Frequency",
    barmode="group",
    width=600
)
age.show()

sex = px.histogram(data, x="sex")
sex.update_layout(
    xaxis_title="",
    yaxis_title="Frequency",
    barmode="group",
    width=600
)
sex.show()

anatomy = px.histogram(data, x="anatom_site_general_challenge")
anatomy.update_layout(
    xaxis_title="",
    yaxis_title="Frequency",
    barmode="group",
    width=600
)
anatomy.show()

benign_malignant = px.histogram(data, x="benign_malignant")
benign_malignant.update_layout(
    xaxis_title="",
    yaxis_title="Frequency",
    barmode="group",
    width=600
)
benign_malignant.show()
```
